```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.amazon.in/s?k=flagship+phones&crid=13SYL6SC92Z3O&sprefix=flagship+phone%2Caps%2C310&ref=nb_sb_noss_1'
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
soup.prettify()
```

```python
products = soup.find_all('span',{'class':'a-size-medium a-color-base a-text-normal'})
```

```python
products
```

```python
prices = soup.find_all('span',{'class':'a-price-whole'})
```

```python
prices
```

```python
products_new = []
for i in products:
  products_new.append(i.get_text())
```

```python
prices_new = []
for i in prices:
  prices_new.append(i.get_text())
```

```python
products_new
```

```python
prices_new
```

```python
import pandas as pd
new_data_frame = pd.DataFrame({'Product Names':products_new,'prices':prices_new})
```

```python
new_data_frame
```

```python
new_data_frame.to_csv('products.csv')
```
